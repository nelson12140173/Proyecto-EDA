```python
# Importar librer√≠as
import pandas as pd
import numpy as np
import os
import joblib
```

```python
# Cargar pipeline final
ruta_pipeline = Path(r"C:\Users\52358965\Proyecto\notebooks\final_model_pipeline.pkl")
pipeline_final = joblib.load(ruta_pipeline)
print("‚úÖ Pipeline final cargado correctamente.")
```

```python
# Cargar dataset de testing
ruta_test = Path(r"C:\Users\52358965\Proyecto\data\raw\Base_Test.xlsx")

if not ruta_test.exists():
    raise FileNotFoundError(f"‚ùå No se encontr√≥ el archivo de testing en: {ruta_test}")

df_test = pd.read_excel(ruta_test)
print("‚úÖ Dataset de testing cargado correctamente.")
print("Shape:", df_test.shape)
print(df_test.head())
```

```python
# Asegurar que no haya columna objetivo
if 'Valores' in df_test.columns:
    df_test = df_test.drop(columns=['Valores'])
```

```python
# 4. Generar predicciones
preds = pipeline_final.predict(df_test)

df_pred = df_test.copy()
df_pred["Prediccion_Valores"] = preds

print("\n‚úÖ Predicciones generadas:")
print(df_pred.head())
```

```python
# Guardar CSV con predicciones
salida_csv = Path(r"C:\Users\52358965\Proyecto\notebooks\predicciones_testing.csv")
df_pred.to_csv(salida_csv, index=False, encoding="utf-8-sig")

print(f"\nüìÅ Archivo generado con √©xito: {salida_csv}")
```
