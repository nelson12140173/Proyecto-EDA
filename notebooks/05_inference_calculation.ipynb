```python
# Importar librerías
import pandas as pd
import numpy as np
import os
import joblib
```

```python
# Cargar pipeline final
from pathlib import Path
import joblib

ruta_pipeline = Path(r"C:\Users\MERBI10\Downloads\Tareas\Proyecto\Proyecto-EDA\notebooks\final_model_pipeline.pkl")
pipeline_final = joblib.load(ruta_pipeline)

print("Pipeline final cargado correctamente.")
```

```python
# Cargar dataset de testing
from pathlib import Path
import pandas as pd

ruta_test = Path(r"C:\Users\MERBI10\Downloads\Tareas\Proyecto\Proyecto-EDA\data\raw\Base_Test.xlsx")

if not ruta_test.exists():
    raise FileNotFoundError(f"No se encontró el archivo de testing en: {ruta_test}")

df_test = pd.read_excel(ruta_test)

print("Dataset de testing cargado correctamente.")
print("Shape:", df_test.shape)
df_test.head()
```

```python
# Asegurar que no haya columna objetivo
if 'Valores' in df_test.columns:
    df_test = df_test.drop(columns=['Valores'])
```

```python
# 4. Generar predicciones
preds = pipeline_final.predict(df_test)

df_pred = df_test.copy()
df_pred["Prediccion_Valores"] = preds

print("\nPredicciones generadas:")
print(df_pred.head())
```

```python
# Guardar CSV con predicciones
from pathlib import Path

salida_csv = Path(r"C:\Users\MERBI10\Downloads\Tareas\Proyecto\Proyecto-EDA\notebooks\predicciones_testing.csv")

df_pred.to_csv(salida_csv, index=False, encoding="utf-8-sig")

print(f"Archivo generado con éxito: {salida_csv}")
```
